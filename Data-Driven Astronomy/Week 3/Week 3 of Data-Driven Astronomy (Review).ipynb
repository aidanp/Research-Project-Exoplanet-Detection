```python
# init.sql
CREATE TABLE Star (
  kepler_id INTEGER NOT NULL,
  t_eff INTEGER NOT NULL,
  radius FLOAT NOT NULL,
  PRIMARY KEY (kepler_id)
);

CREATE TABLE Planet (
  kepler_id INTEGER NOT NULL,
  koi_name VARCHAR(20) NOT NULL,
  kepler_name VARCHAR(20),
  status VARCHAR(20) NOT NULL,
  period FLOAT NOT NULL,
  radius FLOAT NOT NULL,
  t_eq INTEGER NOT NULL,
  PRIMARY KEY (koi_name)
);


COPY Star (kepler_id, t_eff, radius) FROM 'stars2.csv' CSV;

COPY Planet (kepler_id, koi_name, kepler_name, status, period, radius, t_eq) FROM 'planets.csv' CSV;

# my code
SELECT ROUND(AVG(p.t_eq), 1), MIN(s.t_eff), MAX(s.t_eff)
FROM Star AS s
JOIN Planet AS p USING (kepler_id)
WHERE s.t_eff > (
  SELECT AVG(s.t_eff)
  FROM Star AS s
  );
```

This query takes the average planet t_eq values rounded to the nearest tenth, the minimum star t_eff values, and the maximum star t_eff values, but combined with the planet data filling all empty kepler_id with NULL values as the JOIN function does. The output is only when the star t_eff values are greater than the AVG star t_eff value. The subquerry has to be created, because aggregate functions such as AVG or COUNT can't be used in the WHERE function.

```python
# init.sql
CREATE TABLE Star (
  kepler_id INTEGER NOT NULL,
  t_eff INTEGER NOT NULL,
  radius FLOAT NOT NULL,
  PRIMARY KEY (kepler_id)
);

CREATE TABLE Planet (
  kepler_id INTEGER NOT NULL,
  koi_name VARCHAR(20) NOT NULL,
  kepler_name VARCHAR(20),
  status VARCHAR(20) NOT NULL,
  period FLOAT NOT NULL,
  radius FLOAT NOT NULL,
  t_eq INTEGER NOT NULL,
  PRIMARY KEY (koi_name)
);


COPY Star (kepler_id, t_eff, radius) FROM 'stars2.csv' CSV;

COPY Planet (kepler_id, koi_name, kepler_name, status, period, radius, t_eq) FROM 'planets.csv' CSV;

# my code

SELECT p.koi_name, p.radius, s.radius
FROM Planet AS p
INNER JOIN Star AS s USING(kepler_id)
WHERE s.radius IN (
   SELECT radius 
   FROM Star
   ORDER BY radius DESC
   LIMIT 5
 );
```

This query takes the planet koi_name values, the planet radius values, and the star radius values taking data from the planet data and joining that data with the star data using kepler_id. The output is only the greatest 5 star radius values after being ordered in descending order by star radius values.
